This notebook is a sandbox for exploring 3D rotations.

```python
# this is the best 3D rotations import in Python
import numpy as np
from scipy.spatial.transform import Rotation as R
```

```python
# Let's look at a random 3D rotation:
rot = R.random().as_matrix()

# rot @ rot.T 
rot
```

scipy has its own internal representation of 3D rotations called `scipy.spatial.transform._rotation.Rotation`. (Under the hood, it uses a slick representation called a unit quarternion. The math is very slick, but not for today.)

For our uses, we can display the same rotation in different ways:

1. as a rotation matrix: `.as_matrix()`
1. as a unit quaternion: `.as_quat()`
1. as a rotation vector: `as_rotvec()`

The first and third options are worth our attention today.

```python
# camera - to - world : NO
# world - to - camera : YES
z = np.array([0, 0, 1])
rot @ z
```

```python
# this is camera - to - world
rot.T
```

```python
rot @ rot.T
```
